```python
import numpy as np
import pandas as pd
import os

#For test train split
from sklearn.model_selection import train_test_split
#To create a confusion matrix
from sklearn.metrics import confusion_matrix
#Loading library
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Reading the data: 
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
X = iris[['sepallength','sepalwidth','petallength','petalwidth']]
y = iris['species']
```

```python
(X_train , X_test, y_train , y_test) = train_test_split(X,y,test_size=0.3,stratify=y, random_state = 100)
```

```python
#instantiate learning model (k=3)

knn = KNeighborsClassifier(n_neighbors=3)  #3NN

#fitting the model
knn.fit(X_train , y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
c= confusion_matrix(y_test , y_pred)
c
```

```python

# IMPORTING AND READING THE CARS FILE
cars=pd.read_csv("cars.csv")

cars
```

```python
cars2 = cars.drop(['Car','Origin'],axis =1)
#file_array = np.array( file.drop(['sex','country'],axis=1) )
cars2
```

```python
cars2 = cars2.apply(lambda x: ( x - np.mean(x)) / np.std(x))
```

```python
X = cars2[['Weight','Horsepower','Displacement']]
y = cars2['MPG']
```

```python
(X_train , X_test, y_train , y_test) = train_test_split(X,y,test_size=0.3, random_state = 4)
```

```python
knn = KNeighborsRegressor(n_neighbors=3)  #3NN
knn.fit(X_train , y_train)      #fitting the model
y_pred = knn.predict(X_test)
```
